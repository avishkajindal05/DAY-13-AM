## 1. Exact Prompt Used

Write a Python function that takes a Pandas DataFrame and generates an automated data quality report including:

shape

dtypes

missing values percentage

duplicate rows

unique value counts per column

basic statistics

Return the report as a dictionary and also print a formatted summary.

```python

# 2. AI Generated Python Code
import pandas as pd

def data_quality_report(df: pd.DataFrame):
    
    if df.empty:
        print("Warning: DataFrame is empty.")
    
    report = {}

    report["shape"] = df.shape

    report["dtypes"] = df.dtypes.astype(str).to_dict()

    missing_percent = (df.isnull().sum() / len(df) * 100) if len(df) > 0 else 0
    report["missing_percent"] = missing_percent.to_dict()

    report["duplicate_rows"] = int(df.duplicated().sum())

    report["unique_counts"] = df.nunique(dropna=False).to_dict()

    report["single_value_columns"] = [
        col for col in df.columns if df[col].nunique(dropna=False) <= 1
    ]

    report["basic_stats"] = df.describe(include="all").to_dict()

    report["memory_mb"] = df.memory_usage(deep=True).sum() / (1024**2)

    print("\n===== DATA QUALITY SUMMARY =====")
    print(f"Shape: {report['shape']}")
    print(f"Duplicate Rows: {report['duplicate_rows']}")
    print(f"Memory Usage (MB): {report['memory_mb']:.4f}")
    
    print("\nColumn Data Types:")
    for col, dtype in report["dtypes"].items():
        print(f"{col}: {dtype}")
    
    print("\nMissing Values (%):")
    for col, val in report["missing_percent"].items():
        print(f"{col}: {val:.2f}%")
    
    print("\nUnique Values Per Column:")
    for col, val in report["unique_counts"].items():
        print(f"{col}: {val}")
    
    if report["single_value_columns"]:
        print("\nColumns With Single Unique Value (Potentially Useless):")
        for col in report["single_value_columns"]:
            print(f"- {col}")
    
    return report
```

3. Testing the Function
Test 1 – Clean DataFrame

```python
import pandas as pd

clean_df = pd.DataFrame({
    "name": ["Alice","Bob","Charlie","David"],
    "age": [25,30,35,40],
    "salary": [50000,60000,70000,80000],
    "department": ["HR","IT","Finance","IT"]
})

report_clean = data_quality_report(clean_df)
```

4. Critical Evaluation 

The generated function successfully produces a structured data quality report and prints a readable summary. It correctly calculates the DataFrame shape, column data types, missing value percentages, duplicate rows, unique counts per column, and descriptive statistics. Additionally, it includes a useful metric for memory consumption using df.memory_usage(deep=True), which is important when working with large datasets in data engineering or analytics workflows.

The function also detects columns with a single unique value, which is valuable because such columns often represent useless features in machine learning or analytics pipelines. Identifying them early helps avoid unnecessary noise in models or visualizations.

Regarding edge cases, the function partially handles them. If the DataFrame is empty, it prints a warning. However, missing value percentage calculation could still behave unexpectedly if the DataFrame has zero rows, so safeguards were added to avoid division by zero. Columns containing only null values are still handled properly because nunique(dropna=False) counts nulls.

Potential improvements include adding detection of high-cardinality categorical columns, identifying highly imbalanced features, reporting potential outliers, and highlighting columns with extremely high missing percentages (for example >50%). The report could also include correlation analysis and automatic visualization generation. Exporting the report as JSON or Markdown would further improve usability for automated reporting pipelines.
